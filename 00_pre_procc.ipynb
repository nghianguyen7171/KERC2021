```python
import os
train_dir = ('/home/nghia/PR_LAB/KERC_Chall/kerc2021/KERC21Dataset/KERC21Dataset/preprocessed/train')

dct = {}
for f in os.listdir(train_dir):
    ID = f.split('_')[0]
    f_name = f.split('_')[1].replace('.npy', '')
    #print(f_name)
    if ID not in dct.keys():
        dct[ID] = []
    fea = np.load(os.path.join(train_dir,f))
    dct[ID].extend(fea)
```

```python
import pandas as pd
features = [dct[k] for k in dct.keys()]
new_tab = []
for x in features:
    tmp = [[y] for y in x]
    new_tab.append(tmp)
TAB = {"ID": dct.keys(),
       "Fea": features
}
TAB = pd.DataFrame(TAB)
TAB
```

```python
dct2 = {}
for x in features:
    count = 0
    for y in x:
        col = str(count)
        if col not in dct2.keys():
            dct2[col] = []
        dct2[col].append(y)
        count += 1
dct2.keys()
```

```python
import pandas as pd

TAB2 = {"ID": dct.keys()}
for k in dct2.keys():
    TAB2[k] = dct2[k]
#t = np.transpose(TAB)
TAB2 = pd.DataFrame(TAB2)
TAB2
```
